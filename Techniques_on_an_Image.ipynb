```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('images/input.jpg')
cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Histogram

```python
hist = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()

plt.hist(img.ravel(), 256, [0,256]);
plt.show()

c = ['b', 'g', 'r']

for i, col in enumerate(c):
    histogram = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histogram, color=col)
plt.show()
```

## Drawing Images

* A line on black Square

```python
sq = np.zeros((512,512,3), np.uint8)

cv2.line(sq, (0,0), (511,511), (255,127,0), 2)
cv2.imshow('Blue line on black Square', sq)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

* A Rectangle filled with color

```python
back = np.zeros((512,512,3), np.uint8)

cv2.rectangle(back, (100,100), (300,300), (127,57,127), -1)
cv2.imshow('Rectangle filled with color', back)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

* Circle

```python
back_grd = np.zeros((512,512,3), np.uint8)

cv2.circle(back_grd, (250,250), 100, (255,255,255), -1)
cv2.imshow('Circle filled with color', back_grd)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
